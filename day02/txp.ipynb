```python
import torch
```

```python
t1 = torch.arange(12).reshape(3,4)
t2 = torch.randint(0, 10, (3, 4))
print(f't1Xt2是\n{t1*t2}')
t2 = t2.view(4,3)
print(f't2:\n{t2}, shape:{t2.shape}')
print(f't1@t2是\n{t1@t2}')
```

```python
t1 = torch.randint(0, 10, (3, 4)).to(torch.float32).mean()
print(f't1:\n{t1}, shape:{t1.shape}')
t2 = torch.randint(0, 10, (3, 4)).to(torch.float32)
print(f't2:\n{t2}, shape:{t2.shape}')
t3= t2.mean(dim=0)
print(f't3:\n{t3}, shape:{t3.shape}')
```

```python
t1 = torch.randint(0, 10, (3, 4))
print(f't1:\n{t1}, shape:{t1.shape}')
# 获取第二个轴,索引为1的所有数据
t2 = t1[:,1]
print(f't2:\n{t2}, shape:{t2.shape}')
# 获取第二个轴 索引为1大于5的元素对应的第一个轴的数据
t2 = t1[t1[:,1]>5]
print(f't3:\n{t2}, shape:{t2.shape}')
#
```

```python
t1 = torch.randint(0, 10, (3, 4))
t2 = torch.randint(0, 10, (3, 4))
t1= t1.reshape(-1,3)
t2= t2.reshape(-1,3)
print(f't1:\n{t1}, t2:\n{t2}')
t3 = torch.cat([t1, t2], dim=0)
print(f't3:\n{t3}, shape:{t3.shape}')
```

```python
# 创建一个形状为(3)的随机张量w
w = torch.randint(0, 10, (3,), requires_grad=True, dtype=torch.float32)
print(f'w:\n{w}, shape:{w.shape}')
loss = w**2 +20
loss.mean().backward()
print(f'w.grad:\n{w.grad}, shape:{w.grad.shape}')
w.data = w.data - 0.01*w.grad
print(f'w:\n{w}, shape:{w.shape}')
```
